# 1. Imports

```python
import os
from dotenv import load_dotenv
import pandas as pd
from sqlalchemy import create_engine, text
from urllib.parse import quote_plus
```

# 2. Connection

```python
load_vars = load_dotenv("../src/.env")
if load_vars:
    print("Variáveis de ambiente carregadas com sucesso")
else:
    print("Arquivo .env não encontrado!")
```

```python
connection_string = (
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=tcp:127.0.0.1,1433;"
    f"DATABASE={os.environ['DB_NAME']};"
    f"UID={os.environ['DB_USER']};"
    f"PWD={os.environ['DB_PASSWORD']};"
    "TrustServerCertificate=yes;"
    "Encrypt=no;"
)
```

```python
connection_url = f"mssql+pyodbc:///?odbc_connect={quote_plus(connection_string)}"
```

```python
engine = create_engine(connection_url)
```

# 3. Testando querys

```python
def run_query(sql, params=None, engine=engine):
    """Executa um SELECT e retorna um dataframe."""
    try:
        df = pd.read_sql(text(sql), engine, params=params)
        return df
    except Exception as e:
        raise Exception(f"Erro ao executar query: {e}\nSQL: {sql}")
```

```python
run_query("""
    SELECT *
    FROM ihms
    WHERE id_linha_producao = :id
""", {"id": 1})
```

```python
df = run_query("""
    SELECT *
    FROM ihms
    WHERE id_linha_producao = :id
""", {"id": 1})
```
