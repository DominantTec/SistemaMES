# 1. Imports

```python
import pyodbc
import socket
```

# 2. Testar se a porta está aberta localmente

```python
server_ip = "127.0.0.1"
port = 1433  # troque se sua instância estiver em outra porta

sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
sock.settimeout(2)

print(f"Testando porta {port} ...")

try:
    sock.connect((server_ip, port))
    print("✅ Porta aberta!")
except Exception as e:
    print("❌ Porta fechada:", e)

sock.close()
```

# 3. Testar conexão ODBC no SQL local

```python
driver = "{ODBC Driver 17 for SQL Server}"

server = "127.0.0.1"      # ou NOME-DO-PC
port = 1433               # ou a porta real
database = "IHM_Testes_2"
username = "TesteElectron"           # use exatamente como você loga no SSMS
password = "Bruno@2026"

connection_string = (
    f"DRIVER={driver};"
    f"SERVER={server},{port};"
    f"DATABASE={database};"
    f"UID={username};"
    f"PWD={password};"
    "TrustServerCertificate=yes;"
    "Encrypt=no;"
)

print(connection_string)

try:
    conn = pyodbc.connect(connection_string, timeout=5)
    print("🎉 CONECTADO COM SUCESSO AO SQL LOCAL!")
except Exception as e:
    print("❌ ERRO AO CONECTAR:", e)
```

# 4. Testar uma consulta rápida
