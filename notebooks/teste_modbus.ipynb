# 1. Imports

```python
from pymodbus.client import ModbusTcpClient
```

# 2. Connection

```python
client = ModbusTcpClient(host="192.168.11.89", port=502, timeout=10)
```

```python
if not client.connect():
    print("Não conectou no servidor Modbus")
else:
    rr = client.read_holding_registers(address=0, count=1, slave=1)
    print(rr)
    client.close()
```
